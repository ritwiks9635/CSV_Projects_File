```python
X = data.drop(["fetal_health"], axis = 1)
y = data["fetal_health"]
```

```python
cols = list(X.columns)
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns = cols)
X.head()
```

```python
X.describe()
```

```python
plt.figure(figsize=(20,10))
sns.boxenplot(data = X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import precision_score, recall_score, confusion_matrix, classification_report, accuracy_score, f1_score
from sklearn import metrics
from sklearn.metrics import roc_curve, auc, roc_auc_score
```

```python
pipeline1 = Pipeline([("lr", LogisticRegression(random_state = 42))])
pipeline2 = Pipeline([("dtc", DecisionTreeClassifier(random_state = 42))])
pipeline3 = Pipeline([("rfc", RandomForestClassifier())])
pipeline4 = Pipeline([("svc", SVC())])
pipe = [pipeline1, pipeline2, pipeline3, pipeline4]
```

```python
for x in pipe:
  x.fit(X_train, y_train)
```

```python
pipe_dict = {0: 'Logistic Regression', 1: 'Decision Tree', 2: 'RandomForest', 3: "SVC"}
```

```python
cv_results_accuracy = []
for i, model in enumerate(pipe):
    cv_score = cross_val_score(model, X_train,y_train, cv=10 )
    cv_results_accuracy.append(cv_score)
    print("%s: %f " % (pipe_dict[i], cv_score.mean()))
```

```python
print(pipeline3.score(X_train, y_train))
print(pipeline3.score(X_test, y_test))
```

```python
pred = pipeline3.predict(X_test)
```

```python
parameters = {
    'n_estimators': [100,150, 200, 500,700,900],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,6,8,12,14,16],
    'criterion' :['gini', 'entropy'],
    'n_jobs':[-1,1,None]
}
```

```python
grid_cv = GridSearchCV(estimator = RandomForestClassifier(), param_grid = parameters, cv = 5)
grid_cv.fit(X_train, y_train)
```
